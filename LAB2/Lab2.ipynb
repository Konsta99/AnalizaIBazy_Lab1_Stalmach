Zadanie zostało wykonane na danych drinks.csv

```python
import pandas as pd
```

```python
drinks = pd.read_csv("drinks.csv")

drinks
```

```python
drinks2 = pd.melt(drinks,
                       ["country"],
                       var_name="drink servings",
                       value_name="litres of alcohol")

drinks2.to_csv('drinks2.csv', index=False)

drinks2.head(20)
```

Sorted values by country

```python
drinks3 = drinks2.sort_values(by=['country'])

drinks3.to_csv('drinks3.csv', index=False)

drinks3.head(20)
```

Sorted values by litres of alcohol

```python
drinks4 = drinks2.sort_values(by=['litres of alcohol'])

drinks4.to_csv('drinks4.csv', index=False)

drinks4.head(20)
```
